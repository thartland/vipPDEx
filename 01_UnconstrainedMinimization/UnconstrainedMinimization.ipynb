# Energy Minimization

In this example we solve the following minimization problem

*Find* $u^\star \in H^1_0(\Omega)$ *such that*

$$ u^\star = \operatorname*{argmin}_{u \in H^1_0(\Omega)} J(u), $$

For the energy functional

$$ J(u) := \frac{1}{2} \int_\Omega k(u) \nabla u \cdot \nabla u \, dx - \int_\Omega f\,u \, dx,$$

where

$$k(u) = k_1 + k_2 u^2, $$

and $k_{1}>0$ and $k_{2}\geq 0$. As will be demonstrated, the optimality conditions for this problem are nonlinear when $k_{2}>0$ and so in general an iterative "outer" optimization procedure (e.g., gradient descent or Newton's method) must be applied to determine the minimizer $u^\star$.

## Necessary optimality condition (Euler-Lagrange condition)

Let $J_{u}(u)(\tilde{u})$ denote the first variation of $J(u)$ with respect to $u$ in *direction* $\tilde{u}$, i.e.

$$J_{u}(u)(\tilde{u}) := \left. \frac{\mathrm{d}}{\mathrm{d} \varepsilon} J(u + \varepsilon \tilde{u})\right|_{\varepsilon=0}.$$

The first-order necessary condition for $u^\star$ to be a minimizer of $J$ is that $J(u^\star)(\tilde{u})$ is zero for all admissable variations/directions $\tilde{u}$:

$$ J_{u}(u^\star)(\tilde{u}) = 0 \quad \forall \tilde{u} \in H_0^1(\Omega).$$

### Weak form:

To obtain the weak form of the above necessary condition, we first expand the term $J(u + \varepsilon \tilde{u})$ as

$$J(u + \varepsilon \tilde{u}) = \frac{1}{2} \int_\Omega [k_1 + k_2(u + \varepsilon\tilde{u})^2] \nabla(u + \varepsilon \tilde{u})\cdot\nabla(u + \varepsilon \tilde{u}) dx - \int_\Omega f\,(u+\varepsilon\tilde{u}) dx.$$

We then differentiate

$$
\frac{\mathrm{d}}{\mathrm{d}\varepsilon}J(u+\varepsilon\tilde{u})
= \frac{1}{2}\int_{\Omega}\left[
2\,k_{2}\tilde{u}(u+\varepsilon\tilde{u})
\nabla(u+\varepsilon\tilde{u})\cdot\nabla
(u+\varepsilon\tilde{u})+
(k_{1}+k_{2}(u+\varepsilon\tilde{u})^{2})
2\nabla\tilde{u}\cdot\nabla(u+\varepsilon\tilde{u})
\right]dx-\int_{\Omega}f\,\tilde{u}\,dx,
$$

and set $\varepsilon$ to $0$

$$
\left. \frac{\mathrm{d}}{\mathrm{d}\varepsilon}J(u+\varepsilon\tilde{u})\right|_{\varepsilon=0}=
\frac{1}{2}\int_{\Omega}\left[
2\,k_{2}\tilde{u}\,u
\nabla u \cdot\nabla
 u+
(k_{1}+k_{2}\,u^{2})
2\nabla\tilde{u}\cdot\nabla u
\right]dx-\int_{\Omega}f\,\tilde{u}\,dx
$$

$$
J_{u}(u)(\tilde{u})=\int_{\Omega}\left[
k_{2}\tilde{u}\,u
\nabla u \cdot\nabla
 u+
(k_{1}+k_{2}\,u^{2})
\nabla\tilde{u}\cdot\nabla u
\right]dx-\int_{\Omega}f\,\tilde{u}\,dx.
$$

We then write the weak form of the first-order necessary condition for optimality as

*Find* $u\in H_0^1(\Omega)$ *such that*

$$ \int_\Omega \left[k_2 u\, \tilde{u} \nabla u \cdot \nabla u + (k_1 + k_2u^2)\nabla \tilde{u}\cdot \nabla u\right] dx = \int_\Omega f \tilde{u} dx \quad \forall \tilde{u} \in H_0^1(\Omega).$$

### Strong form:
To obtain the strong form, we invoke Green's first identity and write

$$ \int_\Omega \left[k_2 u \nabla u \cdot \nabla u -  \nabla \cdot [(k_1 + k_2u^2) \nabla u] \right] \tilde{u} dx + \int_{\partial \Omega} [(k_1 + k_2u^2) \nabla u]\cdot n \tilde{u} ds = \int_\Omega f \tilde{u} dx \quad \forall \tilde{u} \in H_0^1(\Omega).$$

Since $\tilde{u}$ is arbitrary in $\Omega$ and $\tilde{u} = 0$ on $\partial \Omega$, the strong form of the non-linear boundary problem reads

$$ - \nabla \cdot [(k_1 + k_2u^2) \nabla u] + k_2 u \nabla u \cdot \nabla u = f \quad {\rm in} \; \Omega; $$
$$ u = 0 \quad {\rm on} \; \partial\Omega.$$

## Infinite-dimensional Newton's Method

The infinite-dimensional Newton's method reads

*Given the minimizer estimate* $u_{k}$, *of the energy functional* $J(u)$, *find* $\hat{u}_{k} \in H^1_0(\Omega)$ *such that*

$$ J_{u,u}(u_{k})(\tilde{u}, \hat{u}_{k}) = -J_{u}(u_{k})(\tilde{u}) \quad \forall \, \tilde{u} \in H_0^1(\Omega).$$

*Update the solution using the Newton direction* $\hat{u}_{k}$
$$ u_{k+1} = u_k + \alpha\,\hat{u}_{k}.$$

Where $\alpha>0$ is an appropriately chosen step-length.

### Hessian

To derive the weak form of the Hessian

$$
J_{u,u}(u)(\tilde{u},\hat{u})=\left. \frac{\mathrm{d}}{\mathrm{d}\varepsilon}J_{u}(u+\varepsilon\hat{u})(\tilde{u})\right|_{\varepsilon=0},
$$

we first expand the term $J_{u}(u +\varepsilon \hat{u})(\tilde{u})$ as

$$J_{u}(u+\varepsilon\hat{u})(\tilde{u}) = \int_\Omega \left[k_2 (u+\varepsilon\hat{u}) \tilde{u} \nabla (u+\varepsilon\hat{u}) \cdot \nabla (u+\varepsilon\hat{u}) + (k_1 + k_2(u+\varepsilon\hat{u})^2)\nabla \tilde{u}\cdot \nabla (u+\varepsilon\hat{u}) \right] dx - \int_\Omega f \tilde{u} dx.$$

Then, after some simplification, we obtain

$$J_{u,u}(u)(\tilde{u}, \hat{u}) = \int_\Omega \left[k_2\hat{u}\tilde{u}\nabla u \cdot \nabla u + 2k_2 u \tilde{u} \nabla \hat{u} \cdot \nabla u + 2k_2 u \hat{u} \nabla \tilde{u} \cdot \nabla u + (k_1 + k_2u^2) \nabla \tilde{u} \cdot \nabla \hat{u} \right] dx. $$

### Weak form of Newton step:

*Given* $u \in H_0^1(\Omega)$, *find* $\hat{u} \in H^1_0(\Omega)$ *such that*

$$
\int_\Omega \left[k_2\hat{u}\tilde{u}\nabla u \cdot \nabla u + 2k_2 u \tilde{u} \nabla \hat{u} \cdot \nabla u + 2k_2 u \hat{u} \nabla \tilde{u} \cdot \nabla u + (k_1 + k_2u^2) \nabla \tilde{u} \cdot \nabla \hat{u} \right] dx =
-
\int_\Omega \left[k_2 u \tilde{u} \nabla u \cdot \nabla u + (k_1 + k_2u^2)\nabla \tilde{u}\cdot \nabla u\right] dx - \int_\Omega f \tilde{u} dx \quad \forall\tilde{u}\in H^1_0(\Omega)
$$


The solution is then updated using the Newton direction $\hat{u}$

$$ u^{\rm new} = u + \alpha \hat{u}.$$

Here $\alpha$ denotes a relaxation parameter (back-tracking/line-search) used to achieve global convergence of the Newton method.

### Strong form of the Newton step

$$ - \nabla \cdot \left[ (k_1 + k_2 u^2) \nabla \hat{u}\right] + 2k_2u\nabla\hat{u}\cdot\nabla u - \nabla\cdot(2k_2 u \hat{u} \nabla u) + k_2 \hat{u} \nabla u \nabla u = \nabla \cdot\left[(k_1 + k_2 u^2)\nabla u \right] - k_2 u \nabla u\cdot \nabla u + f \quad {\rm in} \, \Omega.$$
$$ \hat{u} = 0 \quad {\rm on} \, \partial \Omega. $$

```python
# Import FEniCSx
import dolfinx as dl
import ufl

from mpi4py import MPI
from petsc4py import PETSc

# Import the package of mathematical functions
import math
import numpy as np

# Enable plotting inside the notebook
import matplotlib.pyplot as plt
import pyvista as pv
pv.set_jupyter_backend('static')
```

## 2. Define the mesh and finite element spaces

We construct a triangulation (mesh) $\mathcal{T}_h$ of the computational domain $\Omega := [0, 1]^2$ with `nx` elements in the *x*-axis direction and `ny` elements in the *y*-axis direction.

On the mesh $\mathcal{T}_h$, we then define the finite element space $V_h \subset H^1(\Omega)$ consisting of globally continuous piecewise linear functions and we create a function $u \in V_h$.

By denoting by $\lbrace\phi_i(x)\rbrace_{i=1}^{{\rm dim}(V_h)}$ the finite element basis for the space $V_h$ we have
$$ u(x) = \sum_{i=1}^{{\rm dim}(V_h)} \boldsymbol{u}_i \phi_i(x), $$
where $\boldsymbol{u}\in\mathbb{R}^{{\rm dim}(V_h)}$ contains the coefficients for the finite element expansion of $u(x)$.

Finally we define two special types of functions: the `TestFunction` $\hat{u}$ and the `TrialFunction` $\tilde{u}$. These special types of functions are used by `FEniCS` to generate the finite element vectors and matrices which stem from the first and second variations of the energy functional $J(u)$.

```python
comm = MPI.COMM_WORLD
rank = comm.Get_rank()
nprocs = comm.Get_size()
iAmRoot = rank == 0
# Define the finite element mesh. The mesh size h is 1/nx
nx = 16
ny = nx
    
mesh = dl.mesh.create_unit_square(comm, nx, ny, dl.mesh.CellType.triangle)
cells, types, coords = dl.plot.create_vtk_mesh(mesh, mesh.topology.dim)
grid = pv.UnstructuredGrid(cells, types, coords)

# Create plotter
window_size=[400,400]
plotter = pv.Plotter(window_size=window_size)
plotter.add_mesh(grid, show_edges=True)
plotter.view_xy()
plotter.add_title('Finite Element Mesh', font='courier', color='w',
                   font_size=10)
plotter.show()

# Define the finite element space V_h as the space of piecewise linear functions on the elements of the mesh.
degree = 1
Vh = dl.fem.FunctionSpace(mesh, ("CG", degree))
uh  = dl.fem.Function(Vh)

print("Number of dofs", uh.vector.size)
```

## 3. Define the energy functional

We now define the energy functional
$$ J(u) = \frac{1}{2} \int_\Omega (k_1 + k_2u^2) \nabla u \cdot \nabla u dx - \int_\Omega f\,u dx.$$

The parameters $k_1$, $k_2$ and the forcing term $f$ are defined in `FEniCSx` using the keyword `Constant`. To define coefficients that are space dependent one should use the keyword `Function` and then interpolate using a lambda function.

The Dirichlet boundary condition
$$ u = 0 \quad {\rm on} \; \partial\Omega$$
is imposed using the `dirichletbc` class.

To construct this object we need to provide

- the finite element space `Vh`

- the value `u_0` of the solution at the Dirichlet boundary. `u_0` can either be a `Constant` or an `Function` object.

- the lambda function `marker` that defines on which part of $\partial \Omega$ we want to impose such condition.

```python
k1 = dl.fem.Constant(mesh, PETSc.ScalarType(0.05))
k2 = dl.fem.Constant(mesh, PETSc.ScalarType(1.0))
f  = dl.fem.Constant(mesh, PETSc.ScalarType(1.0))

u_bc = dl.fem.Constant(mesh, PETSc.ScalarType(0.0))

facet_dim = mesh.topology.dim-1
facets_D = dl.mesh.locate_entities_boundary(mesh, dim=facet_dim, \
                                        marker=lambda x: np.logical_or(np.logical_or(np.isclose(x[0], 0.0), \
                                        np.logical_or(np.isclose(x[1], 0.0), np.isclose(x[0], 1.0))),
                                        np.isclose(x[1], 1.0)))
dofs_D = dl.fem.locate_dofs_topological(V=Vh, entity_dim=facet_dim, entities=facets_D)
bcs = [dl.fem.dirichletbc(u_bc, dofs_D, Vh)]

Jform = PETSc.ScalarType(0.5)*(k1 + k2*uh*uh)*ufl.inner(ufl.grad(uh), ufl.grad(uh))*ufl.dx - f*uh*ufl.dx
J = dl.fem.form(Jform)
```

## 4. First variation

The weak form of the first variation reads

$$J_{u}(u)(\tilde{u}) = \int_\Omega \left[k_2 u \tilde{u} \nabla u \cdot \nabla u + (k_1 + k_2u^2)\nabla \tilde{u}\cdot \nabla u\right] dx - \int_\Omega f \tilde{u} dx .$$

We use a **finite difference check** to verify that our derivation is correct.
More specifically, we consider a function
$$ u_0 = x(x-1)y(y-1) \in H^1_0(\Omega), $$
and verify that for a random direction $u_{\text{dir}} \in H^1_0(\Omega)$ we have
$$ r_{1}(\varepsilon) := \left| \frac{J(u_0 + \varepsilon\, u_{\text{dir}}) - J(u_0)}{\varepsilon} - J_{u}(u_0)(u_{\text{dir}})\right| = \mathcal{O}(\varepsilon).$$
Neglecting discretization errors we have
$$
J_{u}(u_{0})(u_{\text{dir}})=\boldsymbol{g}(u_{0})^{\top}\boldsymbol{u}_{\text{dir}},
$$
where
$$
u_{\text{dir}}(x)=\sum_{i=1}^{\text{dim}(V_{h})}\boldsymbol{u}_{\text{dir},i}\phi_{i}(x), \\
\boldsymbol{g}(u_{0})_{i}=J_{u}(u_{0})(\phi_{i}),\,\,\,1\leq i\leq \text{dim}(V_{h}).
$$


In the figure below we show in a loglog scale the value of the residual $r$ of the finite difference approximation of $J_{u}(u_{0})(u_{\text{dir}})$ as a function of $\varepsilon$. We observe that $r$ decays linearly for a wide range of values of $\varepsilon$, however we notice an increase in the error for extremely small values of $\varepsilon$ due to numerical stability and finite precision arithmetic.

**NOTE:** To compute the first variation we can also use the [automatic differentiation](https://fenicsproject.org/olddocs/dolfin/2017.2.0/python/programmers-reference/fem/formmanipulations/derivative.html#dolfin.fem.formmanipulations.derivative) of variational forms capabilities of the Unified Form Language `ufl` and write

`gradform = ufl.derivative(Jform, uh)`

```python
u_tilde = ufl.TestFunction(Vh)
u_hat   = ufl.TrialFunction(Vh)
gradform = (k2*uh*u_tilde)*ufl.inner(ufl.grad(uh), ufl.grad(uh))*ufl.dx + \
       (k1 + k2*uh*uh)*ufl.inner(ufl.grad(uh), ufl.grad(u_tilde))*ufl.dx - f*u_tilde*ufl.dx
grad = dl.fem.form(gradform)

u0 = dl.fem.Function(Vh)
u0.interpolate(lambda x: x[0]*(x[0]-1)*x[1]*(x[1]-1))
uh.vector[:] = u0.vector[:]

n_eps = 32
epss = 1e-2*np.power(2., -np.arange(n_eps))
fdgrad_residual = np.zeros(n_eps)
J0 = dl.fem.assemble_scalar(J)
grad0 = dl.fem.petsc.assemble_vector(grad)
dl.fem.petsc.set_bc(grad0, bcs)
grad0.assemble()


u_dir =  dl.la.create_petsc_vector(Vh.dofmap.index_map, Vh.dofmap.index_map_bs)
u_dir[:] = np.random.randn(len(u_dir[:]))

dl.fem.petsc.set_bc(u_dir, bcs)

dir_grad0 = grad0.dot(u_dir)

for i, eps in enumerate(epss):
    uh.vector[:] = u0.vector[:] # uh = u0
    uh.vector.axpy(eps, u_dir) # uh = uh + eps[i]*dir
    Jplus = dl.fem.assemble_scalar(J)
    fdgrad_residual[i] = abs( (Jplus - J0)/eps - dir_grad0 )

plt.loglog(epss, fdgrad_residual, "-ob", label="Error Grad")
plt.loglog(epss, (.5*fdgrad_residual[0]/epss[0])*epss, "-.k", label=r"First Order, $\propto\varepsilon$")
plt.title("Finite difference check of the first variation")
plt.xlabel(r"$\varepsilon$")
plt.ylabel(r"$r_{1}(\varepsilon)$, finite-difference error")
plt.legend(loc = "upper left")
plt.show()
```

## 5. Second variation

The weak form of the second variation reads

$$J_{u,u}(u)(\tilde{u}, \hat{u}) := \left. \frac{d}{d\varepsilon} J_{u}(u+\varepsilon\hat{u})(\tilde{u})\right|_{\varepsilon=0} = 
\int_\Omega \left[k_2\tilde{u}\hat{u}\nabla u \cdot \nabla u + 2k_2 u \hat{u} \nabla \tilde{u} \cdot \nabla u + 2k_2 u \tilde{u} \nabla \hat{u} \cdot \nabla u + (k_1 + k_2u^2) \nabla \hat{u} \cdot \nabla \tilde{u} \right] dx. $$

As before, we verify that for at $u=u_0$, a random direction $u_{\text{dir}} \in H^1_0(\Omega)$ and $\tilde{u}\in H^1_0(\Omega)$ we have
$$ \left| \frac{J_{u}(u_0 + \varepsilon u_{\text{dir}})(\tilde{u}) - J_{u}(u_0)(\tilde{u})}{\varepsilon} - J_{u,u}(u_0)(\tilde{u}, u_{\text{dir}})\right| = \mathcal{O}(\varepsilon),$$

$$ r_{2}(\varepsilon) := \left\| \frac{\boldsymbol{g}(u_{0}+\varepsilon\, u_{\text{dir}}) - \boldsymbol{g}(u_{0})}{\varepsilon} - \boldsymbol{H}(u_{0})\boldsymbol{u}_{\text{dir}} \right\|_{2} = \mathcal{O}(\varepsilon),$$

where $\boldsymbol{g},\boldsymbol{\hat{u}}\in\mathbb{R}^{{\rm dim}(V_{h})}$, $\boldsymbol{H}\in\mathbb{R}^{{\rm dim}(V_{h})\,\times\, {\rm dim}(V_{h})}$ are such that

$$
\left(\boldsymbol{g}(u)\right)_{i} = J_{u}(u)(\phi_{i}) \quad 1\leq i\leq {\rm dim}(V_{h}),
$$

$$
\left(\boldsymbol{H}(u)\right)_{i,j} = J_{u,u}(u)(\phi_{i},\phi_{j}) \quad 1\leq i,j\leq {\rm dim}(V_{h}), 
$$


$$
\left(\boldsymbol{H}(u)\boldsymbol{\hat{u}}\right)_{i}=
J_{u,u}(u)(\phi_{i},\hat{u}) \quad 1\leq i \leq {\rm dim}(V_h).
$$


In the figure below we show in a loglog scale the value of $r$ as a function of $\varepsilon$. As before, we observe that $r$ decays linearly for a wide range of values of $\varepsilon$, however we notice an increase in the error for extremely small values of $\varepsilon$ due to numerical stability and finite precision arithmetic.

**NOTE:** To compute the second variation we can also use the automatic differentiation capabilities of the Unified Form Language `ufl` and write

`Hform = ufl.derivative(gradform, uh)`

```python
Hform = k2*u_hat*u_tilde*ufl.inner(ufl.grad(uh), ufl.grad(uh))*ufl.dx + \
     PETSc.ScalarType(2.0)*(k2*uh*u_tilde)*ufl.inner(ufl.grad(u_hat), ufl.grad(uh))*ufl.dx + \
     PETSc.ScalarType(2.0)*k2*u_hat*uh*ufl.inner(ufl.grad(uh), ufl.grad(u_tilde))*ufl.dx + \
     (k1 + k2*uh*uh)*ufl.inner(ufl.grad(u_tilde), ufl.grad(u_hat))*ufl.dx
H = dl.fem.form(Hform)

uh.vector[:] = u0.vector[:]
H_0 = dl.fem.petsc.assemble_matrix(H, bcs)
H_0.assemble()
H_0udir   = H_0.createVecLeft()
diff_grad = H_0.createVecLeft()


H_0.mult(u_dir, H_0udir)
fdH_residuals = np.zeros(n_eps)

for i, eps in enumerate(epss):
    uh.vector[:] = u0.vector[:]
    uh.vector.axpy(eps, u_dir)
    grad_plus = dl.fem.petsc.assemble_vector(grad)
    dl.fem.petsc.set_bc(grad_plus, bcs)
    diff_grad.zeroEntries()
    diff_grad.axpy(1., grad_plus)
    diff_grad.axpy(-1., grad0)
    diff_grad.scale(1./eps)
    diff_grad.axpy(-1, H_0udir)
    fdH_residuals[i] = diff_grad.norm(2)
    
plt.figure()    
plt.loglog(epss, fdH_residuals, "-ob", label="Error Hessian")
plt.loglog(epss, (.5*fdH_residuals[0]/epss[0])*epss, "-.k", label="First Order")
plt.title("Finite difference check of the second variation")
plt.xlabel(r"$\varepsilon$")
plt.ylabel(r"$r_{2}(\varepsilon)$, finite-difference error")
plt.legend(loc = "upper left")
plt.show()
```

## 6. The infinite dimensional Newton Method

The infinite dimensional Newton step reads

*Given an objective minimizer estimate* $u_k \in H_0^1(\Omega)$, *find* $\hat{u}_{k} \in H^1_0(\Omega)$ *such that*
$$ J_{u,u}(u_k)(\tilde{u}, \hat{u}_{k}) =  - J_{u}(u_k)(\tilde{u}) \quad \forall \tilde{u}\in H^1_0(\Omega)$$
Update the solution $u_{k+1}$ using the Newton direction $\hat{u}_{k}$
$$ u_{k+1} = u_k + \alpha \,\hat{u}_{k}.$$

Here, for simplicity, we choose $\alpha$ equal to 1. In general, to guarantee global convergence of the Newton method the parameter $\alpha$ should be appropriately chosen (e.g. *backtracking* or *line search*).

The linear systems to compute the Newton directions are solved using the conjugate gradient (CG) with algebraic multigrid preconditioner with a fixed tolerance. In practice, one should solve the Newton system inexactly by early termination of CG 
iterations via EisenstatWalker (to prevent oversolving) and Steihaug (to avoid negative curvature) criteria.

In the output below, for each iteration we report the number of CG iterations, the value of the energy functional, the norm of the gradient, and the inner product between the gradient and the Newton direction $J_{u}(u_k)(\hat{u}_{k})$.

In the example, the stopping criterion is relative norm of the gradient $$\frac{
||\boldsymbol{g}(u_{k})||_{2}}{||\boldsymbol{g}(u_0)||_{2}} \leq \tau,$$
where $\tau$ is the relative tolerance `rgradtol`. However robust implementation of the stopping criterion should also monitor the quantity $\boldsymbol{g}(u_{k})^{\top}\boldsymbol{\hat{u}}_{k}$.

```python
uh.vector[:] *= 0.
rgradtol = 1.e-9
max_iter = 10
total_cg_its = 0
J0 = dl.fem.assemble_scalar(J)
g0 = dl.fem.petsc.assemble_vector(grad)
dl.fem.petsc.set_bc(g0, bcs)

grad0norm = g0.norm(2)
tol = grad0norm * rgradtol
uhat = dl.fem.Function(Vh)

print ("{0:3} {1:3}  {2:15} {3:15} {4:15}".format(
      "It", "cg its", "    J    ", "(g, uhat)", "||g||l2") )

for i in range(max_iter):
    Hn = dl.fem.petsc.assemble_matrix(H, bcs)
    Hn.assemble()
    gn = dl.fem.petsc.assemble_vector(grad)
    dl.fem.petsc.set_bc(gn, bcs)
    gradnorm = gn.norm(2)
    if gradnorm < tol:
        print("\nNorm of the gradient at minimizer estimate", gradnorm )
        print("Value of the energy functional at minimizer estimate", Ji)
        print("Converged in {0:d} Newton iterations and {1:d} total conjugate-gradient iterations.".format(i, total_cg_its))
        break
    opts = PETSc.Options()
    opts["ksp_type"] = "cg"
    opts["ksp_rtol"] = min(0.5, np.sqrt(gradnorm/grad0norm))
    opts["pc_type"]  = "gamg"
    # Create PETSc Krylov solver
    solver = PETSc.KSP().create(mesh.comm)
    solver.setFromOptions()

    # Set matrix operator
    solver.setOperators(Hn)
    
    # Solve linear system Hn uhat = -gn
    solver.solve(-gn, uhat.vector)
    # record number of CG-iterations for linear system solution
    cg_its = solver.its
    total_cg_its += cg_its
    # uh = uh + uhat (better to use backtracking line search for robustness)
    uh.vector.axpy(1.0, uhat.vector)
    
    
    
    Ji = dl.fem.assemble_scalar(J)

    print ("{0:3d} {1:2d}       {2:1.2e}      {3:1.2e}       {4:1.2e}".format(
      i, cg_its, Ji, -gn.dot(uhat.vector), gn.norm(2)) )   


grid.point_data["u"] = uh.x.array.real
plotter = pv.Plotter(window_size=window_size)
plotter.add_mesh(grid, show_edges=True, show_scalar_bar=True, scalars="u")
plotter.view_xy()
plotter.add_title('Optimizer', font='courier', color='w', font_size=10)
plotter.show()
```

Alternatively, we can use the nonlinear solver that is provided in `FEniCSx`. Here one specify the nonlinear form that one would like to solve `gradform == 0`, the function that one is solving for `uh`, the essential (Dirichlet) conditions `bcs` that `uh` must satisfy, the Jacobian `J` of `gradform` (which is here the Hessian `Hform` and one specify parameters to choose the numerical algorithms that are used to approximately determine a stationary point of the energy objective functional $J(u)$. 

```python
uh.vector[:] *= 0.
optProblem = dl.fem.petsc.NonlinearProblem(gradform, uh, bcs=bcs, J=Hform)
solver = dl.nls.petsc.NewtonSolver(comm, optProblem)
solver.convergence_criterion = "incremental"
solver.rtol = rgradtol

# We can customize the linear solver used inside the NewtonSolver by
# modifying the PETSc options
ksp = solver.krylov_solver
opts = PETSc.Options()
option_prefix = ksp.getOptionsPrefix()
opts["ksp_type"] = "cg"
opts["pc_type"]  = "gamg"
ksp.setFromOptions()
NewtonIts, converged = solver.solve(uh)
assert(converged)

final_g = dl.fem.petsc.assemble_vector(grad)
dl.fem.petsc.set_bc(final_g, bcs)

print("Number of Newton iterations = {0:d}".format(NewtonIts))
print("Norm of the gradient at minimizer estimate = {0:1.2e}".format(final_g.norm(2)))
print("Value of the energy functional at minimizer estimate = {0:1.2e}".format(dl.fem.assemble_scalar(J) ))


grid.point_data["u (PETSc solver)"] = uh.x.array.real
plotter = pv.Plotter(window_size=window_size)
plotter.add_mesh(grid, show_edges=True, show_scalar_bar=True, scalars="u (PETSc solver)")
plotter.view_xy()
plotter.add_title('Optimizer (PETSc solver)', font='courier', color='w', font_size=10)
plotter.show()
```
